# WHO Dataset Column Reference
## Alcohol per Capita Consumption (EE6F72A)

| Column | Explanation |
|---|---|
| `IND_ID` | Internal numeric ID for the indicator |
| `IND_CODE` | Short code for the indicator (e.g. `SA_0000001688`) |
| `IND_UUID` | Unique identifier for the indicator (e.g. `EE6F72A`) |
| `IND_PER_CODE` | Code for the time period type (e.g. annual) |
| `DIM_TIME` | The year of the data (2000–2019) |
| `DIM_TIME_TYPE` | Type of time measurement (e.g. `YEAR`) |
| `DIM_GEO_CODE_M49` | Country code in UN M49 standard (numeric) |
| `DIM_GEO_CODE_TYPE` | Type of geographic entity (e.g. `COUNTRY`) |
| `DIM_PUBLISH_STATE_CODE` | Publication status of the data point |
| `IND_NAME` | Full name of the indicator |
| `GEO_NAME_SHORT` | Country name (e.g. `Germany`) |
| `DIM_SEX` | Sex disaggregation (Male / Female / Both) |
| `RATE_PER_CAPITA_N` | The actual numeric value (litres of pure alcohol) |
| `RATE_PER_CAPITA_NL` | Lower bound of the confidence interval |
| `RATE_PER_CAPITA_NU` | Upper bound of the confidence interval |

> **Key columns for analysis:** `GEO_NAME_SHORT`, `DIM_TIME`, `DIM_SEX`, and `RATE_PER_CAPITA_N`
