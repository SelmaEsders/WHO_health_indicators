```python
# Import ..
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

```

```python
path = r"C:\Users\ThinkPad\Desktop\Mastermind\WHO\WHO_health_indicators\WHS2025_DATADOWNLOAD.csv"
df = pd.read_csv(path)
df.info()
```

```python
df.columns
```

```python
# check if indicator name is connected to indicator code

# 1. Check if the mapping is one-to-one (each name maps to exactly one code, and vice versa)
name_to_code = df.groupby('IndicatorName')['IndicatorCode'].nunique()
code_to_name = df.groupby('IndicatorCode')['IndicatorName'].nunique()

print("Names with multiple codes:", (name_to_code > 1).sum())
print("Codes with multiple names:", (code_to_name > 1).sum())


# check column "disaggregation"
df["Disaggregation"].nunique()
```
